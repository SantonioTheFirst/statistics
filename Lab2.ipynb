```python
import numpy as np
import scipy.stats as stats
```

```python
def F(u, l) : return 1 - np.exp(-(u * l))
  

def D_k(y, k, n): return np.maximum(y - (k - 1) / n, k / n - y)
    
    
def D(x, l, n):
    x_sorted = np.sort(x)
    D_n = 0
    for i in range(n):
        t = D_k(F(x_sorted[i],l), i, n)
        D_n = t if D_n < t else D_n
    return D_n


def K_test(x, z, l):
    n = x.shape[0]; D_n = D(x, l, n); criteria = np.sqrt(n) * D_n
    print(f'l = {l}\nn = {n}\nD_n(x) = {D_n}\nะัะธัะตัะธะน = {criteria}\nz = {z}')
    print('ะะฐะฝะฝัะต ะฝะต ะฟัะพัะธะฒะพัะตัะฐั ะณะธะฟะพัะตะทะต H_0\n'
          if criteria < z else
          'ะะฐะฝะฝัะต ะฟัะพัะธะฒะพัะตัะฐั ะณะธะฟะพัะตะทะต H_0, ะฟัะธะฝะธะผะฐะตะผ ะณะธะฟะพัะตะทั H_1\n')

    
def create_v(x, l, r):
    interval = 0; x_n = F(x, l); left = 0; right = 1.0 / r; v = np.zeros(r)
    for i in range(r):
        U_i = x_n[(left <= x_n) & (x_n < right)]
        v[i] = len(U_i)
        left += 1 / r; right += 1 / r
    return v


def calc_delta(v, p, n, r):
    summ = 0.0
    for i in range(r):
        summ += (v[i] - n * p) ** 2 / (n * p)
    return summ
        
    
def chi2(x, l, g):
    n = len(x); r = np.int64(20 * len(x) / 1000); z = stats.chi2.ppf(1 - g, r - 1); v = create_v(x, l, r)
    p = 1 / r; d = calc_delta(v, p, n, r)
    print(f'\nl = {l}\nn = {n}\nn = {n}\nr = {r}\nะะปะธะฝะฐ ะธะฝัะตัะฒะฐะปะฐ p = {p}\nะะตะปััะฐ = {d}\nz = {z}')
    print('ะะฐะฝะฝัะต ะฝะต ะฟัะพัะธะฒะพัะตัะฐั ะณะธะฟะพัะตะทะต H_0\n'
          if d < z else
          'ะะฐะฝะฝัะต ะฟัะพัะธะฒะพัะตัะฐั ะณะธะฟะพัะตะทะต H_0, ะฟัะธะฝะธะผะฐะตะผ ะณะธะฟะพัะตะทั H_1\n')
    
    
def get_nu_0(v): return v[v == 0].shape[0]
    
    
def empty_boxes(x, l):
    n = len(x); p = 2; r = np.int64(n / p); z = stats.norm.ppf(1 - 0.05)
    criteria = r * np.exp(-p) + z * np.sqrt(r * np.exp(-p) * (1 - (1 + p) * np.exp(-p))); v = create_v(x, l, r)
    nu_0 = get_nu_0(v)
    print(f'\nl = {l}\nn = {n}\nn = {n}\nr = {r}\nะะปะธะฝะฐ ะธะฝัะตัะฒะฐะปะฐ p = {1 / r}\nะะพะป-ะฒะพ ะฟััััั ััะธะบะพะฒ = {nu_0}\nะัะธัะตัะธะน = {criteria}')
    print('ะะฐะฝะฝัะต ะฝะต ะฟัะพัะธะฒะพัะตัะฐั ะณะธะฟะพัะตะทะต H_0\n'
          if nu_0 < criteria else
          'ะะฐะฝะฝัะต ะฟัะพัะธะฒะพัะตัะฐั ะณะธะฟะพัะตะทะต H_0, ะฟัะธะฝะธะผะฐะตะผ ะณะธะฟะพัะตะทั H_1\n')
    
    
def D_p(y, m, l):
    y_sorted = np.sort(y)
    first = y_sorted[0]
    for i in range(m):
        t = ((i + 1) / m) - F(y_sorted[i], l)
        first = t if first < t else first
    return first


def D_m(y, m, l):
    y_sorted = np.sort(y)
    first = y_sorted[0]
    for i in range(m):
        t = F(y_sorted[i], l) - (i / m)
        first = t if first < t else first
    return first


def S_test(x, y, z, l):
    n = x.shape[0]; m = y.shape[0]; D_plus, D_minus = D_p(y, m, l), D_m(y, m, l); D = np.maximum(D_plus, D_minus)
    criteria = z * np.sqrt((1 / n) + (1 / m))
    print(f'\nl = {l}\nn = {n}\nn = {n}\nm = {m}\nD_plus = {D_plus}\nD_minus = {D_minus}\nD = {D}\nะัะธัะตัะธะน = {criteria}')
    print('ะะฐะฝะฝัะต ะฝะต ะฟัะพัะธะฒะพัะตัะฐั ะณะธะฟะพัะตะทะต H_0\n'
          if D < criteria else
          'ะะฐะฝะฝัะต ะฟัะพัะธะฒะพัะตัะฐั ะณะธะฟะพัะตะทะต H_0, ะฟัะธะฝะธะผะฐะตะผ ะณะธะฟะพัะตะทั H_1\n')
    
       
def do_all_stuff(tasks, lambd, n):
    for t, l in zip(tasks, lambd):
#         print('##################')
        solve_task(t, l, n)
    
       
    
tasks = np.array(['Task A:', 'Task B:', 'Test:'])
lambd = np.array([1.0, 1.3, 0.8])
n = np.array([10 ** 3, 10 ** 4, 10 ** 5])
```
